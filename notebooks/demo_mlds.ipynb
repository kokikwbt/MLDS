```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import sys
sys.path.append('..')
import mlds
```

```python
tensor = np.random.rand(100, 20, 20)
ranks = [2, 2]
```

```python
model = mlds.MLDS(initial_state_cov='full',
                  transition_cov='full',
                  observation_cov='diag',
                  transition_offset=False,
                  observation_offset=False)
```

```python
model.fit(tensor, ranks, max_iter=3, tol=1000)
```

```python
for A in model.A:
    sns.heatmap(A, center=0)
    plt.show()
```

```python
for C in model.C:
    sns.heatmap(C, center=0)
    plt.show()
```
